## Seleçôes de frequencias

```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/aluguel_residencial.csv', sep=';')
dados
```

```python
# selecionar somente os imóveis classificados com tipo 'Apartamento'
selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao] # Ao colocar resultados booleanos dentro de uma lista ele retornar somente os True
n1
```

```python
# Frequencia de apartemento 
n1.shape[0] # or len(n1)
```

```python
# Selecione os imóveis classificados com tipo 'Casa' , 'casa condominio',  'casa de vila'
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao]
n2
```

```python
# frequencia do n2
n2.shape[0] 
```

```python
# selecione imóveis com área entre 60 e 100 m de área , incluindo os limites
selecao = (dados['Area'] >= 60) &  (dados['Area'] <= 100) 
n3 = dados[selecao].shape[0]
n3
```

```python
#selecione os imoveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00
selecao = (dados['Quartos'] >=4) & (dados['Valor'] <= 2000)
n4 = dados[selecao].head(15)
n4
```

```python
n4 = n4[['Tipo', 'Quartos', 'Valor']]
n4
```
